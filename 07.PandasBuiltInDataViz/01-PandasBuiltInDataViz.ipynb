# Встроенная визуализация pandas

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

## Данные

```python
df1 = pd.read_csv('df1', index_col=0)
df1.head()
```

```python
df2 = pd.read_csv('df2')
df2.head()
```

## Виды графиков

### Площадь

```python
df2.plot.area(alpha=0.4)
```

### Столбчатая диаграмма

```python
df2.plot.bar()
```

```python
df2.plot.bar(stacked=True)
```

### Гистограмма

```python
df1['A'].plot.hist(bins=50)
```

### Линейный график

```python
df1.plot.line(y='B', figsize=(12, 3), lw=1)
```

### Скаттерплот

```python
df1.plot.scatter(x='A', y='B')
```

Точки на графике можно раскрасить в разные цвета в зависимости от другой колонки

```python
df1.plot.scatter(x='A', y='B', c='C', cmap='coolwarm')
```

Также точки на графике можно сделать разных размеров в зависимости от другой колонки

```python
df1.plot.scatter(x='A', y='B', s=df1['C'] * 200)
```

### Боксплот

```python
df2.plot.box()
```

### Шестиугольный цветовой график

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
df.head()
```

```python
df.plot.hexbin(x='a', y='b', gridsize=25, cmap='Oranges')
```

### KDE

```python
df2['a'].plot.kde()
```

```python
df2.plot.density()
```
