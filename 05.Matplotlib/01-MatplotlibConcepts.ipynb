# Matplotlib

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

## Базовый пример

### Данные

Лучше всего в качестве исходных данных для графиков использовать массивы numpy или датафреймы pandas.

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
x
```

```python
y = x ** 2
y
```

## Базовые команды matplotlib

Создание графика

```python
plt.plot(x, y, 'r') # 'r' означает красный цвет
plt.show()
```

Подпись осей X и Y

```python
plt.plot(x, y, 'r')
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.show()
```

Подпись всего графика 

```python
plt.plot(x, y, 'r')
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.title('График')
plt.show()
```

Несколько графиков на одной основе

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'g*-')
```

## Объектно-ориентированный метод matplotlib

Объектно-ориентированный метод matplotlib заключается в том, что мы создаем некторый объект, а потом модифицируем его,
вызывая его различные методы.

### Введение в объектно-ориентированный метод

Создаем объект - пустую основу

Добавляем на пустую основу оси. В качестве параметров указывается позиция (left, right, width, height)

Рисуем на этих осях график и подписи

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y, 'b')
axes.set_xlabel('Ось X')
axes.set_ylabel('Ось Y')
axes.set_title('График')
```

Пример двух вложенных графиков

```python
# Создаем пустую основу
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

# Большой график
axes1.plot(x, y, 'b')
axes1.set_xlabel('Ось X графика 1')
axes1.set_ylabel('Ось Y графика 1')
axes1.set_title('График 1')

# Маленький график
axes2.plot(y, x, 'r')
axes2.set_xlabel('Ось X графика 2')
axes2.set_ylabel('Ось Y графика 2')
axes2.set_title('График 2')
```

## subplots()

Объект subplots() позволяет более удобно обращаться с осями

```python
# Отличие subplots() заключается в том, что сразу создается основа и графики
fig, axes = plt.subplots()

# Теперь используя объекта графика, нарисуем график
axes.plot(x, y, 'r')
axes.set_xlabel('Ось X')
axes.set_ylabel('Ось Y')
axes.set_title('График')
```

При создании графиков через subplots можно сразу указывать количество строк и столбцов, что разобъет основу на несколько графиков

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

Тогда объект графиков будет массивом

```python
axes
```

Соответственно, по этому массиву можно итерироваться

```python
for ax in axes:
    ax.plot(x, y, 'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')
    
fig
```

Или обращаться к ним по индексу

```python
axes[0].plot(x, y, 'r')
axes[0].set_title('Первый график')

axes[1].plot(y, x, 'b')
axes[1].set_title('Второй график')
fig
```

Для того, чтобы подписи не залазили друг на друга использую фунцию tight_layout

```python
for ax in axes:
    ax.plot(x, y, 'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')
    
fig.tight_layout()
fig
```

## Легенда, подписи и надписи

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

### Заголовок

```python
ax.set_title('Title')
fig
```

### Подписи осей

```python
ax.set_xlabel('Ось X')
ax.set_ylabel('Ось Y')
fig
```

### Легенда


```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x ** 2, label='x**2')
ax.plot(x, x ** 3, label='x**3')
ax.legend()
```

Легенда имеет параметр loc, который указывает, где расположить легенду

```python
ax.legend(loc=1) # upper right corner
ax.legend(loc=2) # upper left corner
ax.legend(loc=3) # lower left corner
ax.legend(loc=4) # lower right corner
```

## Настройки цвета, ширины и типа линий

### Цвета в нотации MatLab

```python
fig, ax = plt.subplots()
ax.plot(x, x ** 2, 'b.-')
ax.plot(x, x ** 3, 'g--')
```

### Цвета через параметр

```python
fig, ax = plt.subplots()

ax.plot(x, x + 1, color="blue", alpha=0.5)
ax.plot(x, x + 2, color="#8B008B")
ax.plot(x, x + 3, color="#FF8C00")
```

### Стили линий и маркеров

Для установки ширины линии используется параметр linewidth или lw. 

```python
fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(x, x + 1, color='red', linewidth=0.25)
ax.plot(x, x + 2, color='red', linewidth=0.50)
ax.plot(x, x + 3, color='red', linewidth=1.00)
ax.plot(x, x + 4, color='red', linewidth=2.00)
```

Для установки стиля используется параметр linestyle или ls.

```python
fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')
```

Можно устанавливать свой стиль

```python
fig, ax = plt.subplots(figsize=(12, 6))

line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10])
```

Маркеры могут также быть разнных стилей: '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...

```python
fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')
```

Можно отдельно устанавливать размер и цвет маркеров

```python
fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green")
```

## Дополнительные настройки графиков

### Масштаб графиков

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].plot(x, x ** 2, x, x ** 3)
axes[0].set_title("Масштаб по умолчанию")

axes[1].plot(x, x ** 2, x, x ** 3)
axes[1].axis('tight')
axes[1].set_title("Масштаб под размер графика")

axes[2].plot(x, x ** 2, x, x ** 3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("Пользовательский масштаб")
```
