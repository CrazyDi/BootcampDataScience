# Категориальные графики

```python
import seaborn as sns
%matplotlib inline

import numpy as np
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

## barplot

По умолчанию график строится по средним значениям

```python
sns.barplot(x='sex', y='total_bill', data=tips)
```

Построим среднеквадратичные отклонения

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

## countplot

Этот график считает только количество появлений

```python
sns.countplot(x='sex', data=tips)
```

## boxplot

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', hue='smoker', data=tips, palette='coolwarm')
```

## volinplot

```python
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', hue='sex', data=tips, palette='Set1')
```

```python
sns.violinplot(x='day', y='total_bill', hue='sex', data=tips, palette='Set1', split=True)
```

## stripplot

```python
sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=False)
```

```python
sns.stripplot(x='day', y='total_bill', hue='sex', data=tips, palette='Set1')
```

```python
sns.stripplot(x='day', y='total_bill', hue='sex', data=tips, palette='Set1', split=True)
```

## swarmplot

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
sns.swarmplot(x='day', y='total_bill', hue='sex', data=tips, palette='Set1', split=True)
```

## Комбинирование разных графиков

```python
sns.violinplot(x="tip", y="day", data=tips, palette='rainbow')
sns.swarmplot(x="tip", y="day", data=tips, color='black', size=3)
```

## factorplot

```python
sns.factorplot(x='sex', y='total_bill', data=tips, kind='bar')
```
